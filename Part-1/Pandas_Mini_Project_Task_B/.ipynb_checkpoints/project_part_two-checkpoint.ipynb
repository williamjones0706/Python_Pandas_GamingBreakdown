```python
# Dependencies
import csv
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
bootcamp_df = pd.read_csv(file, encoding='iso-8859-1', low_memory=False)
bootcamp_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
bootcamp_df = bootcamp_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
bootcamp_df = bootcamp_df.replace({0.0: "No", 1.0: "Yes"})
bootcamp_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = bootcamp_df.loc[bootcamp_df["AttendedBootcamp"] == "Yes"]
attended_bootcamp.count()
```
